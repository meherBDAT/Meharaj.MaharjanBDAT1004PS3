# Question 1
Answer:

```python
#Step 1: Importing the necessary files

import pandas as pd

#Step 2 and 3: importing the dataset and assigning it to a variable called 'users'

users = pd.read_csv(r"https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
```

```python
#Step 4 : discovering the meand age per occupation
meanAgePerOccupation = users.groupby("occupation").age.mean()
meanAgePerOccupation 
```

```python
#Step 5 : discovering the male ratio per occupation and sorting it from the most to the least

males = users.where(users.gender=="M").groupby(["occupation","gender"]).gender.count()
total = users.groupby("occupation").gender.count()
maleRationPerOccupation = (males / total)*100
maleRationPerOccupation.sort_values(ascending=False)
```

```python
#Step 6 : calculating the minimum and maximum ages for each occupation
minAndMax = users.groupby(["occupation"]).age.agg(["min","max"])
minAndMax
```

```python
#Step 7 : calculating mean age for each combination of occupation and sex

meanAge = users.groupby(["occupation","gender"]).age.mean()
meanAge
```

```python
#Step 8 : presenting the percentage of women and men for each occupation

genderPercentage = users.groupby('occupation')['gender'].value_counts(normalize=True)*100
genderPercentage
```

```python
#Step 1: importing the necessary things
import pandas as pd
```

```python
#Step 2: importing the dataset and assigning it to a variable called euro12
euro12 = pd.read_csv(r"https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv",sep=",")
euro12.head()
```

```python
#Step 4 : selecting only the Goal column
goals = euro12.Goals
goals
```

```python
#Step 5 : Showing no. of teams that participated in the Euro2012
teamCount=euro12.groupby("Team").Team.count()
print(f'The no. of teams that participated in Euro2012 was', len(teamCount))
```

```python
#Step 6 : showing no. of columns in the dataset
print("Total number of columns in the dataset are",len(euro12.columns))
```

```python
#Step 7 : viewing only columns teams, yellow cards and red cards and assigning them to dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
pd.DataFrame(discipline)
```

```python
#Step 8 : sorting the teams by red cards, then to yellow cards
euro12[['Team', 'Red Cards', 'Yellow Cards']].sort_values(by=['Red Cards','Yellow Cards'])
```

```python
#Step 9 : calculating the mean yellow cards given per team
euro12[['Team', 'Yellow Cards', 'Red Cards']].groupby("Team")["Yellow Cards"].mean()
```

```python
#Step 10 : filtering teams that scored more than 6 goals.
euro12[euro12.Goals > 6]
```

```python
#Step 11 : selecting the teams that start with G.
euro12[euro12.Team.str.startswith("G")]
```

```python
#Step 12 : selecting the first 7 columns
euro12.iloc[:, :7] 
```

```python
#Step 13 : selecting all columns except the last 3
euro12.iloc[:,:-3]
```

```python
#Step 14 : presenting only the shooting accuracy from England, Italy and Russia


selectedCountries = (euro12['Team'] == 'England') | (euro12['Team'] == 'Italy') | (euro12['Team'] == 'Russia') 

euro12.loc[selectedCountries, ['Team', 'Shooting Accuracy']]
```
