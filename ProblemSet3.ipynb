```python
#Step 1: Importing the necessary files

import pandas as pd

#Step 2 and 3: importing the dataset and assigning it to a variable called 'users'

users = pd.read_csv(r"https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
```

```python
#Step 4 : discovering the meand age per occupation
meanAgePerOccupation = users.groupby("occupation").age.mean()
meanAgePerOccupation 
```

```python
#Step 5 : discovering the male ratio per occupation and sorting it from the most to the least

males = users.where(users.gender=="M").groupby(["occupation","gender"]).gender.count()
total = users.groupby("occupation").gender.count()
maleRationPerOccupation = (males / total)*100
maleRationPerOccupation.sort_values(ascending=False)
```

```python
#Step 6 : calculating the minimum and maximum ages for each occupation
minAndMax = users.groupby(["occupation"]).age.agg(["min","max"])
minAndMax
```

```python
#Step 7 : calculating mean age for each combination of occupation and sex

meanAge = users.groupby(["occupation","gender"]).age.mean()
meanAge
```

```python
#Step 8 : presenting the percentage of women and men for each occupation

genderPercentage = users.groupby('occupation')['gender'].value_counts(normalize=True)*100
genderPercentage
```

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('us-marriages-divorces-1867-2014.csv')

dataset
```

```python
 
dataset.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])
plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014 \n', fontsize = 21)
plt.xlabel('Year')
plt.ylabel('per Capita')

plt.show()
```

# Question 7 
Answer:

```python
dataset2 = (dataset['Year'] == 1900) | (dataset['Year'] == 1950) | (dataset['Year'] == 2000) 
dataset2 = dataset.loc[dataset2, ['Year','Marriages_per_1000','Divorces_per_1000']]

dataset2[{'Marriages_per_1000','Divorces_per_1000'}].plot.bar()
plt.title("Marriages and Divorces per capita in the U.S. between 1900, 1950 and 2000", fontsize = 21)
plt.xlabel('Year')
plt.ylabel('per Capita')

plt.xticks((0, 1, 2), ("1900", "1950", "2000"), rotation = 0)
```
